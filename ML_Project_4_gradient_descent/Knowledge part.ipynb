# In this case, we will use a linear regression model. The specific parameters of the model are the weights and bias. 
# y = w * x + b, or h(x) = w * x + b, where w is the weight, b is the bias, x is the input feature, and y is the output prediction.
# The weights and bias are the parameters that the model will learn during training. 
# The weights are the coefficients of the features, and the bias is the intercept of the model. 
# The goal of training is to find the optimal values of the weights and bias that minimize the loss function.
#################################################################################################################
