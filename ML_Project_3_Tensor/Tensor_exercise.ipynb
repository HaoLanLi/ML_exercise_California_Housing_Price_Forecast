```python
#Scalar, 0-rank tensor/0-dimensional tensor
import numpy as np
x =np.array(5)
print("The value of x:",x)
print("The rank of x:",x.ndim)
print("The data type of x:",x.dtype)
print("The shape of x:",x.shape)
```

```python
#Vector, 1-rank tensor/1-dimensional tensor
y = np.array([1,2,3,4,5]) #This is a 1-dimensional tensor with 5 elements.
print("The value of y:",y)
print("The rank of y:",y.ndim)
print("The data type of y:",y.dtype)
print("The shape of y:",y.shape) #It is 5-dimensional vector.

#Vector dataset in ML
from tensorflow.keras.datasets import boston_housing
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
print("The shape of x_train:",x_train.shape) #It is 404-dimensional vector.
print("The shape of the first sample of x_train:",x_train[0].shape) #It is 13-dimensional vector.
print("The shape of y_train:",y_train.shape) #It is 404-dimensional vector.


```
